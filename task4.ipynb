```python
# Problem 1: Mobile Recharge Validation System

def validate_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount: ₹"))

        if amount < 50:
            print(" Minimum recharge must be ₹50. Try again.")
        elif amount in valid_plans:
            print(" Recharge successful with plan ₹", amount)
            break
        else:
            print(" Invalid plan. Valid plans are:", valid_plans)

# Call function
validate_recharge()
```

```python
# Problem 2: Inventory Reorder Alert System

def check_inventory(products):
    for item, stock in products.items():
        if stock < 15:
            print(f"{item}: Reorder Alert (Stock = {stock})")
        else:
            print(f"{item}: Stock OK (Stock = {stock})")

# Example data
inventory = {
    "Rice": 20,
    "Sugar": 10,
    "Oil": 5,
    "Flour": 18
}

check_inventory(inventory)
```

```python
# Problem 3: Student Result Processing System

def process_result(marks):
    total = 0

    for m in marks:
        total += m

    average = total / len(marks)
    print("Average Marks:", average)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"

# Example
marks_list = [45, 60, 70, 55]
result = process_result(marks_list)
print("Result:", result)
```

```python
# Problem 4: Cab Fare Estimator with Retry

def calculate_fare(distance, peak):
    base_fare = 50
    fare = base_fare + (12 * distance)

    if peak.lower() == "yes":
        fare = fare + (fare * 0.25)

    return fare

# Retry loop
while True:
    km = float(input("Enter distance in km: "))
    peak_hour = input("Is it peak hour? (yes/no): ")

    total_fare = calculate_fare(km, peak_hour)
    print("Estimated Fare: ₹", round(total_fare, 2))

    retry = input("Do you want to calculate again? (yes/no): ")
    if retry.lower() != "yes":
        break
```

```python
# Problem 5: Employee Attendance Eligibility Checker

def check_attendance(attendance):
    present = 0

    for day in attendance:
        if day == "P":
            present += 1

    percentage = (present / len(attendance)) * 100
    print("Attendance Percentage:", percentage)

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

# Example
attendance_list = ["P", "P", "A", "P", "P", "A", "P", "P"]
status = check_attendance(attendance_list)
print("Status:", status)
```

```python
# Problem 6: Password Strength Checker

def check_password(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"

    if len(password) < 8:
        return "Weak Password (Minimum length 8 required)"

    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in special_chars:
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Must contain digit and special character)"

# Example
pwd = input("Enter password: ")
print(check_password(pwd))
```
